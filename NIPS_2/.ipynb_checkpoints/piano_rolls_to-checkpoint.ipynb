```python
#python 2 - pianoroll to midi function
import sys
sys.path
sys.path.append('./midi')

import numpy as np
from midi import utils    

def pr2midi(pianoroll, i=1):
    filename = './piano_rolls/pianoroll_'+ str(i) + '.mid'
    utils.midiwrite(filename, pianoroll, (32, 93), 0.25) 
    # 3rd arg is the start and end notes; 4rth arg is time between sample
    i += 1      
```
